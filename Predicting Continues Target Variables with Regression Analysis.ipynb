## Defining Linear Regression Algorithm

```python
class LinearRegressionGD(object):

    def __init__(self, learning_rate=0.001, iterations=20):
        self.learning_rate = learning_rate
        self.iterations = iterations

    def fit(self, X, y):
        self.theta = np.zeros(1 + X.shape[1])
        self.cost_ = []

        for i in range(self.iterations):
            output = self.net_input(X)
            errors = (y - output)
            
            #Changing the Value of weights after each iterations
            self.theta[1:] += self.learning_rate * np.dot(X.T,errors)
            self.theta[0] += self.learning_rate * errors.sum()
            
            #Computing cost after each iteration
            cost = (errors**2).sum() / 2.0
            self.cost_.append(cost)
        return self

    def net_input(self, X):
        return np.dot(X, self.theta[1:]) + self.theta[0]

    def predict(self, X):
        return self.net_input(X)
```

Let's run the Linear Regression algorithm on 'RM' and 'MEDV'

```python
X = np.array(dataset['RM']).reshape((X.shape[0],1))
y = np.array(dataset['MEDV'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s_x = StandardScaler()
s_y = StandardScaler()
```

```python
X_std = s_x.fit_transform(X)
y_std = s_y.fit_transform(y.reshape((X.shape[0],1)))
y_std = y_std.flatten()
```

```python
lr = LinearRegressionGD()
```

```python
lr.fit(X_std,y_std)
```

Plotting Cost vs Iterations

```python
plt.plot(range(1,lr.iterations + 1),lr.cost_)
plt.ylabel('SSE')
plt.xlabel('Iterations')
```

## Using RANSAC Algorithm

```python
from sklearn.linear_model import RANSACRegressor
```

```python
ransac = RANSACRegressor(LinearRegression(),min_samples=50,residual_threshold=5.0,random_state=0)
```

```python
ransac.fit(X,y)
```

```python
inlier_mask = ransac.inlier_mask_
```

```python
outlier_mask = np.logical_not(inlier_mask)
```

```python
line_X = np.arange(3, 10, 1)
line_y_ransac = ransac.predict(line_X[:, np.newaxis])
```

```python
plt.scatter(X[inlier_mask], y[inlier_mask],
            c='steelblue', edgecolor='white', 
            marker='o', label='Inliers')
plt.scatter(X[outlier_mask], y[outlier_mask],
            c='limegreen', edgecolor='white', 
            marker='s', label='Outliers')
plt.plot(line_X, line_y_ransac, color='black', lw=2)
plt.xlabel('Average number of rooms [RM]')
plt.ylabel('Price in $1000s [MEDV]')
plt.legend(loc='upper left')
```

```python
print("Slope: %0.3f" %ransac.estimator_.coef_[0])
```

```python
print("Intercept: %0.3f" %ransac.estimator_.intercept_)
```

## Evaluating the Linear Regression Model

### Alternative 1: Residual Plots

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset.iloc[:,:-1].values
```

```python
y = dataset.iloc[:,-1].values
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
linear_regression_model = LinearRegression()
```

```python
linear_regression_model.fit(X,y)
```

Predict both X_train and X_test

```python
y_train_pred = linear_regression_model.predict(X_train)
y_test_pred = linear_regression_model.predict(X_test)
```

Since, this is multi variable, we cannot Linear regression line , so we calculate the residual and visualise by plotting residuals. Residual plots are for diagnosing the Linear regression Models.

```python
y_train_residual = y_train_pred - y_train
```

```python
y_test_residual = y_test_pred - y_test
```

```python
plt.scatter(y_train_pred,  y_train_residual,c='k',label='Training data')
plt.scatter(y_test_pred,  y_test_residual,c='r',label='Test data')
plt.xlabel('Predicted values')
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0,xmin=-10,xmax=50,color='green',lw=5)
```

Ideally, Residual should be zero. Larger Deviation from zero means, the prediction varies greatly from actual data. Outliers are also visible from Residual plots

### Alternate 2:  Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
```

```python
print('MSE for training data: %0.3f , MSE for Testing data: %0.3f' %(mean_squared_error(y_train,y_train_pred),mean_squared_error(y_test,y_test_pred)))
```

### Alternate 3: R^2 score

R-squared = Explained variation / Total variation

R-squared is always between 0 and 100%:

0% indicates that the model explains none of the variability of the response data around its mean.
100% indicates that the model explains all the variability of the response data around its mean.
In general, the higher the R-squared, the better the model fits your data.

The R-squared for the regression model on the top is 15%, and for the model below is 85%. When a regression model accounts for more of the variance, the data points are closer to the regression line.

```python
from sklearn.metrics import r2_score
```

```python
print('R^2 for training data: %0.3f , R^2 for Testing data: %0.3f' %(r2_score(y_train,y_train_pred),r2_score(y_test,y_test_pred)))
```

# Modeling Non-Linear Relationship

```python
X = dataset['LSTAT'].values.reshape((X.shape[0],1))
```

```python
y = dataset['MEDV'].values
```

```python
plt.scatter(X,y)
plt.xlabel('% lower status of the population')
plt.ylabel('Price in 1000 $')
```

```python
# First get the regressor
reg_lin = LinearRegression()
reg_quad = LinearRegression()
reg_lin.fit(X,y)
```

```python
# Modeling Non-Linear relationships using PolynomialFeatures : mention the degree as parameter
from sklearn.preprocessing import PolynomialFeatures
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)
reg_quad.fit(X_quad,y)
```

```python
X_fit = np.arange(X.min(), X.max(), 1)[:, np.newaxis] #This features values will  be used for predicting output
```

```python
# Predicting using Linear Regression
y_lin_fit = reg_lin.predict(X_fit)

# Predicting using Non- Linear Regression
y_quad_fit = reg_quad.predict(quadratic.fit_transform(X_fit))
```

```python
# Calculating R^2 square score in both cases
# R^2 Square => Linear
linear_r2_score = r2_score(y, reg_lin.predict(X))

# R^2 Square => Quad
quad_r2_score = r2_score(y, reg_quad.predict(X_quad))
```

```python
plt.scatter(X, y, label='training points', color='lightgray')
plt.plot(X_fit, y_lin_fit, 
         label='linear (d=1), $R^2=%.2f$' % linear_r2_score, 
         color='blue', 
         lw=2, 
         linestyle=':')
plt.plot(X_fit, y_quad_fit, 
         label='quadratic (d=2), $R^2=%.2f$' % quad_r2_score,
         color='red', 
         lw=2,
         linestyle='-')

plt.xlabel('% lower status of the population [LSTAT]')
plt.ylabel('Price in $1000s [MEDV]')
plt.legend(loc='upper right')
```
